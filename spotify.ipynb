### Visualization of the trends and patterns in the DataFrame

#### Top 5 Most Streamed Tracks

```python
#Convert the non-numeric data into numeric values in order to visualize the data
data['streams'] = pd.to_numeric(data['streams'], errors = 'coerce')

#Declare a variable for the 5 largest stream values
topfive = data.nlargest(5,'streams')

#Plot and manipulate the bar graph for the visual representation of the top 5 most streamed tracks
colors = ['#FFB3E6','#FFA07A','#FF69B4','#D5006D','#C71585']
plt.figure(figsize=(10, 4))
sns.barplot(x='streams', y='track_name', data=topfive, hue='streams', palette=colors)
sns.set(style="whitegrid")

plt.title('Spotify 2023 Top 5 Most Streamed Tracks')
plt.xlabel('Number of Streams')
plt.ylabel('Track Name')

plt.show()
```

#### Distribution of Streams

```python
#Plot and manipulate a histogram for the visualization of the distribution of streams
plt.figure(figsize=(10, 5))
sns.histplot(data['streams'], bins=30, kde=True, color='green', label='Streams Distribution')

plt.title('Distribution of Streams')
plt.xlabel('Number of Streams')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

#### Streams vs. BPM (Beats Per Minute)

```python
#Use a scatter plot to compare the stream variable against the beats per minute variable
plt.figure(figsize=(10, 5))
sns.scatterplot(x='bpm', y='streams', data=data, alpha=0.5, color='blue')

plt.title('Streams vs. BPM (Beats Per Minute)')
plt.xlabel('BPM')
plt.ylabel('Number of Streams')

plt.show()
```

#### Streams by Artist Count

```python
#Use a boxplot for accurate representation of the streams by artist count
plt.figure(figsize=(10,5))
sns.boxplot(x='artist_count', y='streams', data=data, color='yellow')

plt.title('Streams by Artist Count')
plt.xlabel('Number of Artists')
plt.ylabel('Number of Streams')

plt.show()
```

#### Tracks Released Over Time

```python
#Use and manipulate a lineplot for tracking the period of time in which track releases are increasing
tracks_every_year = data['released_year'].value_counts().sort_values()
plt.figure(figsize=(10,5))
sns.lineplot(x=tracks_every_year.index, y=tracks_every_year.values, marker='s', color='red')

plt.title('Number of Tracks Released Every Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks Released')

plt.show()
```

#### Correlations between different variables 

```python
#Define a function for converting the variables into numeri data and removing non-numeric characters except for dots and commas.
def clean(column):
    #Make sure the column is treated as string type
    column = column.astype(str) 
    column = column.replace(r'[^\d.,]', '', regex=True)
    column = column.str.replace(',', '.')
    return pd.to_numeric(column, errors='coerce')

#Use the declared function in each of the column
data['danceability_%'] = clean(data['danceability_%'])
data['acousticness_%'] = clean(data['acousticness_%'])
data['valence_%'] = clean(data['valence_%'])
data['energy_%'] = clean(data['energy_%'])
data['instrumentalness_%'] = clean(data['instrumentalness_%'])
data['liveness_%'] = clean(data['liveness_%'])
data['speechiness_%'] = clean(data['speechiness_%'])  # Corrected the column name

# Summarize the values into its correlation with one another
correlation = data[['streams', 'danceability_%', 'acousticness_%', 'valence_%', 'energy_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']].corr()
correlation
```

#### Visualization of the Correlation

```python
#Use of heatmap for the visualization of the correlation of different musical charaacteristics
plt.figure(figsize=(7,5))
sns.heatmap(correlation, annot=True, cmap='Greens', center=0)

plt.title('Correlation of Different Musical Characteristics')

plt.show()
```

#### Streams vs Energy via Scatter Plot

```python
#Use and manipulate a scatterplot for the visualization of the streams variable against the energy variable
plt.figure(figsize=(10, 5))
sns.scatterplot(x='streams', y='energy_%', data=data, alpha=0.5, color='purple', label='Energy Percentage')

plt.title('Streams vs Energy')
plt.xlabel('Number of Streams')
plt.ylabel('Energy %')
plt.legend

plt.show()
```
